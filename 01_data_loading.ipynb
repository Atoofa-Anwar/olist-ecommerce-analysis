```python
print(orders.dtypes)
```

```python
orders['order_purchase_timestamp']        = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_approved_at']               = pd.to_datetime(orders['order_approved_at'])
orders['order_delivered_carrier_date']    = pd.to_datetime(orders['order_delivered_carrier_date'])
orders['order_delivered_customer_date']   = pd.to_datetime(orders['order_delivered_customer_date'])
orders['order_estimated_delivery_date']   = pd.to_datetime(orders['order_estimated_delivery_date'])
```

```python
orders.dtypes
```

```python
#how many days required in delivery
orders['delivery_days'] = (
    orders['order_delivered_customer_date'] - 
    orders['order_purchase_timestamp']
).dt.days
print(orders['delivery_days'].describe())
```

```python
#insights
##3,000 orders kabhi deliver nahi hue-cancelled hue hai  ya lost
 #Average 12 din delivery time lg rha hai-improvement ki zarurat hai
#209 din wala order investigate karna chahiye bohot zyada wqt lg rha outlier hai bohot bda ye isko ab next find kiye
```

```python
# now we find otlier excat 
worst_order = orders[orders['delivery_days'] == 209]
print(worst_order[['order_id', 'order_status', 'order_purchase_timestamp', 
                    'order_delivered_customer_date', 'delivery_days']])
```

```python
orders['order_month'] = orders['order_purchase_timestamp'].dt.month
orders['order_year']  = orders['order_purchase_timestamp'].dt.year
orders['order_day']   = orders['order_purchase_timestamp'].dt.day_name()
print(orders[['order_purchase_timestamp', 
              'order_month', 
              'order_year', 
              'order_day']].head())
```

```python
#merging tables

df_master = pd.merge(orders, 
                     order_items, 
                     on='order_id', 
                     how='left')
```

```python
df_master.head()
```

```python
#now merge the customers to master table
df_master = pd.merge(df_master,
                     customers,
                     on='customer_id',
                     how='left')
```

```python
df_master=pd.merge(df_master,payments,
                   on='order_id',
                   how='left')
```

```python
df_master=pd.merge(df_master,products,
                   on='product_id',
                   how='left')
```

```python
df_master=pd.merge(df_master,sellers,
                   on='seller_id',
                   how='left')
```
