```python
import pandas as pd
import numpy as np 
import matplotlib .pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns',None)
df=pd.read_csv("olist_master.csv")
```

```python
#highly month order
monthly_orders = (df.groupby(['order_year', 'order_month'])
                    .agg(total_orders=('order_id', 'count'))
                    .reset_index())
```

```python
monthly_orders['period'] = (monthly_orders['order_year'].astype(str) 
                            + "-" + 
                            monthly_orders['order_month'].astype(str).str.zfill(2))
```

```python
plt.figure(figsize=(14, 5))
```

```python
plt.plot(monthly_orders['period'], 
         monthly_orders['total_orders'], 
         marker='o', 
         color='steelblue', 
         linewidth=2)
plt.title("Monthly Order Trend — Olist (2016–2018)", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Total Orders")
plt.xticks(rotation=45)
```

```python
peak = monthly_orders.loc[monthly_orders['total_orders'].idxmax()]
low  = monthly_orders.loc[monthly_orders['total_orders'].idxmin()]
```

```python
print(peak)
```

```python
print(low)
```

```python
#inights:
# Highest Orders : November 2017 → 9,111 orders
  #Lowest Orders  : December 2016 → only 1 order
 #December 2016 had only 1 order because Olist had just launched — business was brand new

 #November 2017 was the peak — this is because maybe Brazil has a big shopping event or any other events 
     
#From 2016 to 2017 orders grew massively  from 1 order to 9,111 in just one year

  # Overall business grew very fast — strong upward trend shows Olist was gaining popularity in Brazil

#BUSINESS ACTION:
  #Increase stock and delivery staff before November every year to handle peak demand.
```

```python
# Which category had the most orders?

category_orders = (df.groupby('product_category_name')
                     .agg(total_orders=('order_id', 'count'),
                          total_revenue=('payment_value', 'sum'))
                     .reset_index()
                     .sort_values('total_orders', ascending=False)
                     .head())
```

```python
plt.figure(figsize=(12, 6))

sns.barplot(data=category_orders,
            x='total_orders',
            y='product_category_name')

plt.title("Top 10 Product Categories by Orders", fontsize=14)
plt.xlabel("Total Orders")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```

```python
top=category_orders.iloc[0]
print(top)
```

```python
#insights
#Top Category    : Bed & Bath (cama_mesa_banho)
                    #11,823 orders
                   # Revenue: R$ 17,12,553
#Bed & Bath products are most popular in Brazil— people spend heavily on home essentials
#This category alone generated over 17 lakh in revenue — highest among all categories
##BUSINESS ACTION:
#Focus marketing budget on Bed & Bath category Keep this category always well stocked Offer combo deals in this category to increase average order value
```

```python

# Which payment method customers use the most
payment_counts = df['payment_type'].value_counts()
print(payment_counts)
```

```python
payment_counts.plot(kind='bar', 
                    color='steelblue',
                    figsize=(8, 5))

plt.title("Most Used Payment Methods", fontsize=14)
plt.xlabel("Payment Type")
plt.ylabel("Total Orders")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Credit Card is overwhelmingly popular almost 75% customers pay by credit card
#Boleto is 2nd most used Boleto is a Brazilian cash payment slip, like a challan. Many Brazilians dont have credit cards so they use Boleto at shops to pay cash
#BUSINESS ACTION:
#Always keep Credit Card as primary option Never remove Boleto — many customers depend on it
#Offer discounts on Credit Card payments to increase average order value
```
